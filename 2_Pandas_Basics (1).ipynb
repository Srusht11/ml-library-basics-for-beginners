#  Pandas — The Excel Sheet Inside Python

**Pandas** is a Python library that lets you work with data in **tables** (rows and columns) —
just like an Excel sheet, but with superpowers!

>  **Simple idea:** If NumPy is a calculator, Pandas is like a smart **notebook with tables**
> that can sort, search, and organize your data for you.

##  History — Where did Pandas come from?

- Pandas was created in **2008** by **Wes McKinney** while he was working at a finance company.
- He needed a tool to analyze financial data quickly in Python — nothing like it existed yet, so he built it!
- The name "Pandas" comes from **"Panel Data"**, a term used in statistics for data tables.
- It became open-source (free for everyone) and is now one of the most used libraries in data science.

##  Why do we need Pandas?

Imagine a class register with student names, marks, and attendance. You want to:
- Find the topper
- Find students who scored below 40
- Calculate the class average

Doing this with plain lists is messy. Pandas gives you a **DataFrame** (a table) where you can
do all this in just one line of code!

##  Step 1: Import Pandas

```python
import pandas as pd
print("Pandas is ready! Version:", pd.__version__)
```

##  Step 2: Creating a DataFrame (a Table)

```python
data = {
    "Name": ["Aryan", "Bela", "Chirag", "Diya", "Esha"],
    "Math": [80, 90, 45, 70, 60],
    "Science": [85, 78, 50, 88, 65]
}

df = pd.DataFrame(data)
df
```

##  Step 3: Looking at Your Data

```python
print("First 2 rows:")
print(df.head(2))

print("\nColumn names:", list(df.columns))
print("\nNumber of students:", len(df))
```

###  Step 4: Doing Calculations

```python
df["Total"] = df["Math"] + df["Science"]
print(df)

print("\nClass average in Math:", df["Math"].mean())
print("Topper in Total marks:")
print(df[df["Total"] == df["Total"].max()])
```

##  Step 5: Filtering (Finding specific rows)

```python
# Find students who scored more than 70 in Math
good_in_math = df[df["Math"] > 70]
print(good_in_math)
```

##  Step 6: Sorting

```python
sorted_df = df.sort_values("Total", ascending=False)
print(sorted_df)
```

##  Let's Visualize the DataFrame!

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(6,4))
plt.bar(df["Name"], df["Total"], color="orange")
plt.title("Total Marks per Student")
plt.xlabel("Student")
plt.ylabel("Total Marks")
plt.show()
```
