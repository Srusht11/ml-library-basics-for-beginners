#  Matplotlib — The Art Studio for Data

**Matplotlib** is a Python library used to **draw pictures of data** — like bar charts,
line graphs, and pie charts.

>  **Simple idea:** If your data is a story, Matplotlib is the artist that draws pictures
> to help you understand that story at a glance.

###  History — Where did Matplotlib come from?

- Matplotlib was created by **John D. Hunter**, a neuroscientist, in **2003**.
- He needed a way to visualize brain signal data in Python (similar to a tool called MATLAB, which
  was expensive and not open-source).
- He wanted something **free** that anyone could use — so he built Matplotlib.
- Sadly, John Hunter passed away in 2012, but his creation lives on and is used by millions of people
  every single day. A yearly award, the "John Hunter Excellence in Plotting Award," honors his legacy.

##  Why do we need Matplotlib?

Numbers in a table are hard to understand quickly. But a **picture** tells the story instantly!

Example: "Ravi scored 45, 60, 75, 90 in 4 tests" — as numbers, it's just numbers.
But as a **line graph going up**, you instantly see: "Ravi is improving!" 📈

##  Step 1: Import Matplotlib

```python
import matplotlib.pyplot as plt
print("Matplotlib is ready!")
```

###  Step 2: Your First Line Graph

```python
tests = ["Test 1", "Test 2", "Test 3", "Test 4"]
ravi_scores = [45, 60, 75, 90]

plt.plot(tests, ravi_scores, marker="o", color="green")
plt.title("Ravi's Progress Over 4 Tests")
plt.xlabel("Test")
plt.ylabel("Score")
plt.show()
```

##  Step 3: Bar Chart (comparing things)

```python
subjects = ["Math", "Science", "English", "Hindi"]
marks = [80, 92, 65, 70]

plt.bar(subjects, marks, color=["red", "blue", "green", "orange"])
plt.title("Marks by Subject")
plt.ylabel("Marks")
plt.show()
```

##  Step 4: Pie Chart (parts of a whole)

```python
activities = ["Study", "Play", "Sleep", "Other"]
hours = [4, 2, 8, 10]

plt.pie(hours, labels=activities, autopct="%1.0f%%", colors=["gold","lightgreen","skyblue","pink"])
plt.title("How I Spend My Day (24 hours)")
plt.show()
```

##  Step 5: Scatter Plot (spotting patterns)

```python
study_hours = [1, 2, 3, 4, 5, 6]
scores = [40, 50, 55, 65, 80, 88]

plt.scatter(study_hours, scores, color="purple")
plt.title("Study Hours vs Score")
plt.xlabel("Hours Studied")
plt.ylabel("Score")
plt.show()
```

##  Step 6: Multiple Lines on One Graph

```python
tests = ["Test 1", "Test 2", "Test 3", "Test 4"]
ravi = [45, 60, 75, 90]
meera = [70, 68, 72, 75]

plt.plot(tests, ravi, marker="o", label="Ravi")
plt.plot(tests, meera, marker="s", label="Meera")
plt.title("Ravi vs Meera")
plt.xlabel("Test")
plt.ylabel("Score")
plt.legend()
plt.show()
```
