# 🔢 NumPy — The Super Calculator for Python

**NumPy** stands for **Numerical Python**. Think of it as a *super-fast calculator* that can handle
huge lists of numbers (called **arrays**) all at once, instead of one number at a time.

> 🎈 **Simple idea:** Normal Python is like adding numbers with your fingers.
> NumPy is like having a calculator that adds a *thousand numbers* in one click!

## 📜 History — Where did NumPy come from?

- In the late 1990s, a scientist named **Jim Hugunin** created a tool called **Numeric** to do math in Python.
- Later, another tool called **Numarray** was made to fix some of Numeric's problems.
- In **2005**, a scientist named **Travis Oliphant** combined the best parts of both and created **NumPy**.
- Today, NumPy is used by almost **every** data scientist, engineer, and AI researcher in the world.
- It is completely **free and open-source**, meaning anyone can use and improve it!

## 🤔 Why do we need NumPy?

Imagine you have the marks of 1,000 students and you want to add 5 bonus marks to everyone.

- **Without NumPy:** You'd need a loop that goes through each student one by one. Slow! 🐢
- **With NumPy:** You add 5 to the *entire list* in one line. Super fast! 🚀

NumPy arrays are also more organized — like a neat grid of numbers (rows and columns),
similar to a table in your notebook.

## 🧩 Step 1: Import NumPy
We give it a nickname `np` so we don't have to type the full word every time.

```python
import numpy as np
print("NumPy is ready! Version:", np.__version__)
```

## 🧩 Step 2: Creating an Array
An array is just a list of numbers, but faster and more powerful.

```python
marks = np.array([45, 67, 89, 32, 90])
print("Marks array:", marks)
print("Type of marks:", type(marks))
```

## 🧩 Step 3: Doing Math on the WHOLE Array at Once

```python
bonus_marks = marks + 5
print("Original marks:", marks)
print("After adding 5 bonus marks:", bonus_marks)

doubled = marks * 2
print("Doubled marks:", doubled)
```

## 🧩 Step 4: 2D Arrays (Rows and Columns — like a table!)

```python
# Marks of 3 students in 2 subjects (Math, Science)
table = np.array([
    [80, 90],
    [70, 85],
    [95, 60]
])
print(table)
print("Shape of table (rows, columns):", table.shape)
```

## 🧩 Step 5: Handy Built-in Tools

```python
print("Average marks:", np.mean(marks))
print("Highest mark:", np.max(marks))
print("Lowest mark:", np.min(marks))
print("Sorted marks:", np.sort(marks))
```

## 📊 Let's Visualize an Array!
We'll use another library called Matplotlib (you'll learn it in the next notebook) just to *see* our numbers.

```python
import matplotlib.pyplot as plt

names = ["Aryan", "Bela", "Chirag", "Diya", "Esha"]
plt.figure(figsize=(6,4))
plt.bar(names, marks, color="skyblue")
plt.title("Student Marks (NumPy Array)")
plt.xlabel("Student")
plt.ylabel("Marks")
plt.show()
```

## 🌟 Fun Fact
NumPy is so fast because it's actually written in a language called **C** underneath —
Python just gives it friendly commands, and C does the heavy lifting at lightning speed! ⚡

## ✅ Summary
- NumPy = fast tool for working with numbers and arrays.
- Created in 2005 by Travis Oliphant.
- Arrays can be 1D (a list) or 2D (a table with rows/columns).
- You can do math on entire arrays instantly, no loops needed.
- Almost every AI/ML library (Pandas, TensorFlow, PyTorch, scikit-learn) is built on top of NumPy!
