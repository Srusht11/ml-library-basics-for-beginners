# 🌊 Seaborn — Matplotlib's Stylish Cousin

**Seaborn** is a Python library that also draws graphs (like Matplotlib), but it makes them
look **more beautiful** and adds smart statistics automatically.

> 🎈 **Simple idea:** If Matplotlib is a plain pencil sketch, Seaborn is the same sketch
> but colored in beautifully with extra details, using way less effort from you.

## 📜 History — Where did Seaborn come from?

- Seaborn was created by **Michael Waskom** around **2012-2014** while he was doing
  neuroscience research (studying how brains work).
- He built it **on top of Matplotlib** to make statistical graphs (graphs that show patterns
  and relationships in data) easier and prettier to create.
- The name "Seaborn" comes from a character named **Sam Seaborn** in a TV show called
  *"The West Wing"* — just a fun name the creator liked!

## 🤔 Why do we need Seaborn if we already have Matplotlib?

Matplotlib gives you full control, but you have to write a lot of code for nice-looking graphs.
Seaborn does the hard work for you — nice colors, grids, and even statistics like
"which subject has the most spread in marks" — in just one line!

## 🧩 Step 1: Import Seaborn

```python
import seaborn as sns
import matplotlib.pyplot as plt
print("Seaborn is ready! Version:", sns.__version__)
```

## 🧩 Step 2: A Simple Dataset

```python
import pandas as pd

data = {
    "Subject": ["Math","Science","English","Hindi","Math","Science","English","Hindi"],
    "Marks":   [85, 90, 60, 70, 78, 88, 65, 72],
    "Student": ["Aryan","Aryan","Aryan","Aryan","Bela","Bela","Bela","Bela"]
}
df = pd.DataFrame(data)
df
```

## 🧩 Step 3: A Pretty Bar Chart (one line!)

```python
sns.barplot(data=df, x="Subject", y="Marks", hue="Student")
plt.title("Marks Comparison")
plt.show()
```

## 🧩 Step 4: Box Plot (shows spread of data)
A box plot shows the minimum, maximum, and middle of a group of numbers, all in one picture.

```python
sns.boxplot(data=df, x="Subject", y="Marks")
plt.title("Spread of Marks by Subject")
plt.show()
```

## 🧩 Step 5: Scatter Plot with Style

```python
study_data = pd.DataFrame({
    "Hours_Studied": [1,2,3,4,5,6,7,8],
    "Score": [40,50,55,65,70,80,85,95]
})

sns.scatterplot(data=study_data, x="Hours_Studied", y="Score")
plt.title("Study Hours vs Score")
plt.show()
```

## 🧩 Step 6: Heatmap (colors show patterns)

```python
import numpy as np

marks_table = pd.DataFrame({
    "Math":    [80, 70, 90, 60],
    "Science": [85, 65, 95, 55],
    "English": [70, 75, 88, 60]
}, index=["Aryan","Bela","Chirag","Diya"])

sns.heatmap(marks_table, annot=True, cmap="YlGnBu")
plt.title("Marks Heatmap")
plt.show()
```
