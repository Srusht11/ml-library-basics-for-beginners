# 🧠 TensorFlow — Google's Brain-Building Library

**TensorFlow** is a library made by **Google** for building **Neural Networks** —
computer programs designed to work a little bit like a human brain, made of
connected "neurons" that learn from examples.

> 🎈 **Simple idea:** If Scikit-learn is a toolbox of simple prediction machines,
> TensorFlow lets you build your own "brain" out of building blocks, from scratch.

## 📜 History — Where did TensorFlow come from?

- TensorFlow was created by the **Google Brain team** and released to the public in
  **November 2015**.
- Before TensorFlow, Google used an internal tool called **DistBelief** — TensorFlow was
  built to be faster, more flexible, and open to everyone (not just Google).
- The name comes from **"Tensor"** (a fancy word for a grid of numbers, like an array) +
  **"Flow"** (because data flows through the network step by step).
- It's used by huge companies and researchers worldwide for things like Google Translate,
  Google Photos' face recognition, and speech recognition.

## 🤔 Why do we need TensorFlow?

Some problems are too complex for simple math formulas — like recognizing a handwritten
number, or telling a cat photo from a dog photo. TensorFlow lets us build **Neural
Networks**: layers of connected "neurons" that learn these complex patterns from thousands
of examples.

## 🧩 Step 1: Import TensorFlow

```python
import tensorflow as tf
print("TensorFlow is ready! Version:", tf.__version__)
```

## 🧩 Step 2: What is a Tensor?
A Tensor is just a fancy name for a grid of numbers (like a NumPy array), but TensorFlow
can also run this grid super fast, even on special hardware like GPUs.

```python
scalar = tf.constant(7)                  # a single number
vector = tf.constant([1,2,3])            # a list of numbers
matrix = tf.constant([[1,2],[3,4]])      # a grid (rows & columns)

print("Scalar:", scalar)
print("Vector:", vector)
print("Matrix:\n", matrix)
```

## 🧩 Step 3: Building a Tiny Neural Network
Let's build a simple "brain" that learns the pattern: **y = 2x** (double the input).

```python
import numpy as np

# Training examples: input (x) and correct answer (y)
x_train = np.array([1,2,3,4,5], dtype=float)
y_train = np.array([2,4,6,8,10], dtype=float)

# A tiny neural network with just 1 neuron
model = tf.keras.Sequential([
    tf.keras.layers.Dense(units=1, input_shape=[1])
])

model.compile(optimizer="sgd", loss="mean_squared_error")
print("Model created! It knows nothing yet — let's train it.")
```

## 🧩 Step 4: Training the Model (Let it Learn!)

```python
history = model.fit(x_train, y_train, epochs=200, verbose=0)
print("Training complete!")
print("Final loss (lower = better):", history.history["loss"][-1])
```

## 🧩 Step 5: Make a Prediction

```python
result = model.predict(np.array([10.0]), verbose=0)
print(f"The model thinks 2 x 10 = {result[0][0]:.2f}")
```

## 📊 Let's Visualize the Learning Process

```python
import matplotlib.pyplot as plt

plt.plot(history.history["loss"])
plt.title("How the Model's Mistakes Decreased Over Time")
plt.xlabel("Training Round (Epoch)")
plt.ylabel("Loss (Mistake Level)")
plt.show()
```
